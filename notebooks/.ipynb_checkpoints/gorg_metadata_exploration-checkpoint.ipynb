Scatter plot  

Let's compare SAG assembly length and completeness.

```python
sns.scatterplot(x = 'assembly_length', y = 'completeness_CheckM2', data = df[:100])
```

Hmmm, this doesn't look right, begin in-line troubleshooting! Let's see what data type pandas has assigned to the column 'assembly_length'.

```python
df['assembly_length'].dtype
```

The assembly length column is being treated as an object by pandas.  We'd like to plot it as an integer.  Let's try to change it to an integer:

```python
df['assembly_length'] = [int(i) for i in df['assembly_length']]
```

Oops, an error.  The message suggests that some of the assembly lengths have commas in them, something python would not like. Let's add a step to remove commas before changing the data types within this column to integers.

```python
df['assembly_length'] = [int(i.replace(',','')) for i in df['assembly_length']]
```

Let's try plotting again.

```python
sns.scatterplot(x = 'assembly_length', y = 'completeness_CheckM2', data = df[:100])
```

That looks better!  now lets expand it to the entire dataframe, and color each point by 'gc_content'

```python
sns.scatterplot(x = 'assembly_length', y = 'completeness_CheckM2', hue = 'gc_content', data = df)
```

Ramunas' favorite plot:

```python
df['raw_read_count'] = [int(i.replace(',','')) for i in df['raw_read_count']]
df['plate'] = ["-".join(i.split("-")[:-1]) for i in df['SAG']]
```

```python
df['plate']
```

```python
plates = ['AM-390', 'AM-388']

sdf = df[df['plate'].isin(plates)]

ax = sns.scatterplot(x = 'WGA-X_cp',
                y = 'completeness_CheckM2', 
                data = sdf,
                hue = 'gc_content',
                size = sdf['raw_read_count'] / 10000, 
               palette = 'viridis', 
               alpha = 0.7,   # makes dots transparent
                sizes = (0, 200))

ax.set(xlabel = 'Cp Value')
ax.legend(loc = 'upper right')
ax.set_xlim(1, 6)
plt.show()
```

Plotting FACS data

```python
ax = sns.scatterplot(x = 'X692.40_488',
                y = 'X531.40_488', 
                data = sdf, 
                hue = 'gc_content',
                size = 'assembly_length',
                palette = 'viridis', 
                alpha = 0.7,   # makes dots transparent
                sizes = (10, 400))

ax.set(xscale = 'log',
       yscale = 'log', 
       xlabel = 'Fluorescence at 692 nm, relative units',
       ylabel = 'Fluorescence at 531 nm, relative units')

ax.legend(loc = 'best', ncol = 1)

plt.show()
```
